```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
exercise_df = pd.read_csv('../data/exercise_dataset.csv')
```

```python
# Display basic statistics
print(exercise_df.describe())
```

```python
# Filter numeric columns
numeric_df = exercise_df.select_dtypes(include=['number'])

# Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Exercise Features')
plt.show()
```

```python
# Scatterplot of calories burned vs duration
plt.figure(figsize=(8, 5))
sns.scatterplot(data=exercise_df, x='Duration', y='Calories Burn')  # Or 'Calories' if renamed
plt.title('Calories Burned vs Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(exercise_df['Heart Rate'], kde=True, color='green')
plt.title('Distribution of Heart Rate')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
